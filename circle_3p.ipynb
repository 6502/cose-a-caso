```python
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
import random, time
from IPython.display import clear_output, display, Javascript

def auto_scroll():
    display(Javascript('document.getElementById("site").scrollTo(0, document.getElementById("site").scrollHeight);'))

A = (100, 100)
B = (200, 100)
C = (100, 300)

def dist(p, q):
    dx = p[0] - q[0]
    dy = p[1] - q[1]
    return (dx**2 + dy**2) ** 0.5

def loss(X):
    AX = dist(A, X)
    BX = dist(B, X)
    CX = dist(C, X)
    M = (AX + BX + CX) / 3
    return (M - AX)**2 + (M - BX)**2 + (M - CX)**2

X = ((A[0] + B[0] + C[0]) / 3,
     (A[1] + B[1] + C[1]) / 3)
best = loss(X)

def rnd():
    return (random.random() - 0.5)

plt.figure()
auto_scroll()

start = time.time()

count = 0
accept = 0
while time.time() < start + 10:
    count += 1
    newX = (X[0] + rnd(), X[1] + rnd())
    newloss = loss(newX)
    if newloss < best:
        best = newloss
        X = newX
        accept += 1
    clear_output(wait=True)
    ax = plt.subplots()[1]
    ax.set_xlim(0, 400)
    ax.set_ylim(0, 400)
    ax.set_aspect('equal')
    for (p, name, color) in [(A, "A", "blue"), (B, "B", "blue"), (C, "C", "blue"), (X, "X", "red")]:
        ax.add_patch(Circle((p[0], p[1]), 4, color=color))
        ax.text(p[0] + 10, p[1] - 5, name, fontsize=12, color=color)
    ax.add_patch(Circle((X[0], X[1]), ((dist(X, A) + dist(X, B) + dist(X, C)) / 3), color='red', fill=False))
    plt.title("count=%i/%i   loss=%.9g" % (accept, count, best))
    plt.show()
    auto_scroll()
print(X, best)
```
