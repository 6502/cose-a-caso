```python
import random, time
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from IPython.display import clear_output, display, Javascript

def auto_scroll():
    display(Javascript('document.getElementById("site").scrollTo(0, document.getElementById("site").scrollHeight);'))

def project(cam, x, y, z):
    u = cam[0]*x + cam[1]*y + cam[2]*z + cam[3]
    v = cam[4]*x + cam[5]*y + cam[6]*z + cam[7]
    w = cam[8]*x + cam[9]*y + cam[10]*z + 1
    if w == 0: w = 1
    return u/w, v/w

# xs, ys, x, y, z
data = [
    (278, 584, 0, 1, 0),
    (594, 742, 1, 1, 0),
    (775, 489, 1, 1, 1),
    (816, 202, 1, 0, 1),
    (595, 477, 1, 0, 0),
    (208, 302, 0, 0, 0),
    (461, 72,  0, 0, 1),
]

def loss(cam):
    ee = 0
    for xs, ys, x, y, z in data:
        xx, yy = project(cam, x, y, z)
        ee += (xx - xs)**2 + (yy - ys)**2
    return ee

cam = [1, 0, 0, 500,
       0, 1, 0, 500,
       0, 0, 0]

img = mpimg.imread("/home/agriffini/checkout/talks/pycon2025/cubo2b.jpg")
start = time.time()
last = start - 1
best = loss(cam)

plt.figure()

while time.time() < start + 10:
    cam1 = cam[:]
    cam1[random.randrange(11)] += (random.random() - 0.5)
    e = loss(cam1)
    if e <= best:
        best = e
        cam = cam1

    if time.time() > last + 0.1:
        clear_output(wait=True)
        plt.imshow(img)
        def line(x0, y0, z0, x1, y1, z1):
            xx0, yy0 = project(cam, x0, y0, z0)
            xx1, yy1 = project(cam, x1, y1, z1)
            plt.plot([xx0, xx1], [yy0, yy1], color="blue")
        for i in range(4):
            line(i/3, 0, 0, i/3, 1, 0)
            line(0, i/3, 0, 1, i/3, 0)
            line(1, 0, i/3, 1, 1, i/3)
            line(1, i/3, 0, 1, i/3, 1)
            line(0, 0, i/3, 1, 0, i/3)
            line(i/3, 0, 0, i/3, 0, 1)
        plt.axis("off")
        plt.title("Total error = %.18g" % best)
        plt.show()
        auto_scroll()
        last = time.time()
print(cam, best)
```
